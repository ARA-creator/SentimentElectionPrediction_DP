```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_data(file_path):
  trump_biden_df= pd.read_csv(file_path, encoding= 'utf-8')
  return  trump_biden_df
```

```python
file_path= '/content/drive/MyDrive/preprocessed_datasets/1_labelled_trump_biden_df.csv'
trump_biden_df = load_data(file_path) # Load Biden data
```

```python
trump_biden_df.isnull().sum()
```

```python
missing_values = trump_biden_df[trump_biden_df.isna().any(axis = 1)]
missing_values
```

```python
def drop_missing_preprocessed_tweet(trump_biden_df):
    # Drop rows where 'preprocessed_tweet' is missing
    trump_biden_df = trump_biden_df.dropna(subset=['preprocessed_tweet'])
    return trump_biden_df
```

```python
trump_biden_df = drop_missing_preprocessed_tweet(trump_biden_df)
```

```python
trump_biden_df.isnull().sum()
```

```python
def count_language_values(trump_biden_df):
    return trump_biden_df['language'].value_counts()
```

```python
language_counts = count_language_values(trump_biden_df)
print(language_counts)
```

```python
def plot_language_distribution(language_counts):
    # Define the custom colors: dark blue, light blue, gold, and grey
    colors = ['#00008B', '#0074D9', '#FFD700', '#AAAAAA']

    # Create the bar chart with custom colors
    language_counts.plot(kind='bar', figsize=(10, 6), color=colors)

    # Set the title and labels
    plt.title('Language Distribution')
    plt.xlabel('Language')
    plt.ylabel('Count')

    # Show the plot
    plt.show()
```

```python
# Plot language distribution
plot_language_distribution(language_counts)
```

```python
def filter_english(trump_biden_df):
    # Filter the DataFrame to keep only rows where 'language' is 'en'
    trump_biden_df_filtered = trump_biden_df[trump_biden_df['language'] == 'en']
    return trump_biden_df_filtered
```

```python
trump_biden_df_filtered = filter_english(trump_biden_df)
```
